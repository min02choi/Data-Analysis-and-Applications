- pyplot 위에 figure (그림) 객체를 올리고 그 위에 그래프에 해당하는 축 (axes)를 올려서 그래프를 표현합니다.

```python
import matplotlib.pyplot as plt  # matplotlib 모듈 호출

X = range(100)
Y = range(100)

plt.plot(X, Y)
```

- X 객체와 Y 객체 값 쌍으로 좌표평면 위에 점을 찍고,
- plt.plot 함수는 이 점들을 연결해서 그래프를 그립니다.

```python
import numpy as np # numpy 모듈 호출

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

- pyplot 객체 내부에 있는 하나의 그림 객체 위에 코사인 그래프와 사인 그래프를 그린 예입니다.

**figure와 axes**
- figure은 그래프를 작성하는 밑바탕이 되는 객체입니다.
- axes은 실제로 그래프를 작성하는 공간입니다.

```python
fig, ax = plt.subplots() # (1) figure와 axes 객체 할당

X_1 = range(100)
Y_1 = [np.cos(value)
       for value in X]

ax.plot(X_1, Y_1) # (2) plot 함수를 사용하여 그래프 생성
ax.set(title='cos graph', # (3) 그래프 제목, X축 라벨, Y축 라벨 설정
       xlabel='X',
       ylabel='Y');

plt.show() # (4) 그래프 출력
```

- subplots numpy array의 index로 각 Axes에 접근하여 그래프를 생성합니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,100) # (1) x 값과 y_n 값 생성
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)
y_5 = x ** 2

fig, ax = plt.subplots(2, 2) # (2) 2×2 figure 객체를 생성

ax[0, 0].plot(x, y_1) # (3) 첫 번째 그래프 생성
ax[0, 1].plot(x, y_2) # (4) 두 번째 그래프 생성
ax[1, 0].plot(x, y_3) # (5) 세 번째 그래프 생성
ax[1, 1].plot(x, y_4) # (6) 네 번째 그래프 생성

plt.show()
```

1. plt.subplots 함수 사용 - Axes 2차원 배열 (numpy array)
2. ***plt.subplot 함수 사용***
3. figure 개체의 add_subplot 함수 사용

- subplot 함수는 *(행길이, 열길이, 위치)* 을 입력으로 받습니다.
- 위치는 1부터 *행길이x열길이* 까지 값을 가집니다.

**막대 그래프**
- plt.bar 함수는 막대 그래프를 그리는데 사용됩니다.
- bar 함수의 color 인자를 통해 색상을 지정해줄 수 있습니다.
- bar 함수의 width 인자를 통해 막대의 폭을 지정해줄 수 있습니다.

```python
# (1) 데이터 생성
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

# (2) X 좌표 시작점
X = np.arange(0,8,2)

# (3) 3개의 막대그래프 생성
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)

# (4) X축에 표시될 이름과 위치 설정
plt.xticks(X+0.50, ("A","B","C", "D"))

# (5) 막대그래프 출력
plt.show()
```

- plt.barh 함수를 통해 수평 막대 그래프를 그릴 수 있습니다.
- width 인자 대신 height 인자, xticks 함수 대신 yticks 함수를 이용해야합니다.

```python
plt.barh(X + 0.00, data[0], color = 'b', height = 0.50)
plt.barh(X + 0.50, data[1], color = 'g', height = 0.50)
plt.barh(X + 1.0, data[2], color = 'r', height = 0.50)

plt.yticks(X+0.50, ("A","B","C", "D"))

plt.show()
```

**박스플롯**
- 박스플롯 (boxplot) 또는 박스-위스커 플롯 (Box-Whisker plot)은 사분위수를 시각화하여 데이터의 분포와 밀집 정도를 표현합니다.

1. 최소값 (minimum): Q1 - IQR * 1.5, 정규분포 Q0 근사치
2. **Q1**: The first quartile, 제 1사분위의 수, 25번째 백분위
3. **Q2**: The second quartile, 제 2사분위의 수, 중간값 (median)
4. **Q3**: The third quartile, 제 3사분위의 수, 75번째 백분위
5. 최대값 (maximum): Q3 + IQR * 1.5, 정규분포 Q4

* IQR (Inter Quartile Range) = (Q3-Q1) 근사치
* Outliers (이상치): Q1 - 1.5IQR이하, Q3 + 1.5IQR이상의 값

- 그리드가 생성된 후 맵 (map)을 사용하여 그래프를 만듭니다.
- 각 facet grid에 있는 개별 그래프 영역에 그래프를 집어넣는 구조로, 전체 데이터를 범주형 데이터의 다양한 관점에서 나눠서 볼 수 있습니다.
- 아래는 나눠진 범주에 scatterplot을 그리는 예입니다. 이 때 x는 "total_bill" y는 "tip" 열을 이용해 그립니다.

```python
g = sns.FacetGrid(tips, col="time", row="sex")

g.map(sns.scatterplot, "total_bill", "tip")
```

-  FAcetGrid 위에 seaborn에서 제공하는 다른 그래프 역시 그릴 수 있습니다.

```python
g = sns.FacetGrid(tips, col="time", row="sex")

g.map_dataframe(sns.histplot, x="total_bill")
```

**Plotly**
- 플롯리 (plotly)는 인터렉티브한 시각화가 가능한 파이썬 그래픽 라이브러리 입니다.
- seaborn은 matplotlib의 wrapper 라이브러리인 반면, plotly는 다른 시각화 라이브러리입니다.


```python
import plotly.express as px
df = px.data.iris()
df
```

- Plotly 또한 테스트를 위한 미리 정의된 데이터셋을 제공하며, 데이터 셋의 종류 및 자세한 데이터셋에 대한 설명은 다음 링크를 참고해주세요. https://plotly.com/python-api-reference/generated/plotly.data.html
- iris 데이터셋을 이용하여 간단한 그래프를 생성합니다.

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```

- 생성된 그래프에 마우스 커서를 올리면 데이터를 볼 수 있습니다. (인터랙션 그래프)

**좌표 그래프 (coordinates plot)**
- seaborn에는 없지만 plotly에 존재하는 그래프가 다수 있습니다.
- coordinates plot은 그 예입니다.

```python
fig = px.parallel_coordinates(df, color="species_id",
                              labels={"species_id": "Species", "sepal_width": "Sepal Width", "sepal_length": "Sepal Length", "petal_width": "Petal Width", "petal_length": "Petal Length", },
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
fig.show()
```

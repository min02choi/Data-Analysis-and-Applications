```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,100) # (1) x 값과 y_n 값 생성
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)
y_5 = x ** 2

fig, ax = plt.subplots(2, 2) # (2) 2×2 figure 객체를 생성

ax[0, 0].plot(x, y_1) # (3) 첫 번째 그래프 생성
ax[0, 1].plot(x, y_2) # (4) 두 번째 그래프 생성
ax[1, 0].plot(x, y_3) # (5) 세 번째 그래프 생성
ax[1, 1].plot(x, y_4) # (6) 네 번째 그래프 생성

plt.show()
```

```python
ax1 = plt.subplot(2, 2, 1) # (1) 첫 번째 공간에 axes 생성
plt.plot(x, y_1)
ax2 = plt.subplot(4, 2, 2) # (2) 두 번째 공간에 axes 생성
plt.plot(x, y_2)
ax3 = plt.subplot(424) # (3) 두 번째 공간에 axes 생성
plt.plot(x, y_3, "-.")
ax4 = plt.subplot(413) # (4) 다섯 번째 공간에 axes 생성
plt.plot(x, y_4, "*")
ax5 = plt.subplot(427) # (5) 여섯 번째 공간에 axes 생성
plt.plot(x, y_5, c="red")
ax6 = plt.subplot(428) # (5) 여섯 번째 공간에 axes 생성
plt.plot(x, y_5, c="b", ls="dotted")

plt.show()
```
